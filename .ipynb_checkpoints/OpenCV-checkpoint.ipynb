```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread("demo.png",format='rgb')
```

```python
plt.imshow(img)
```

```python
template = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = template.detectMultiScale(img, 1.1, 5)
```

```python
faces.shape[0]
```

```python
for face in faces:
    x, y ,w, h = face
    img  = cv2.rectangle(img, (x, y), (x+w, y+h), (255,255,0),2)
```

```python
plt.imshow(img)
```

### Face Detect on WebCam

```python
cam = cv2.VideoCapture(0)
while True:
    ret, frame = cam.read()
    
    template = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
    
    if ret == False:
        continue
    
    faces = template.detectMultiScale(frame, 1.1, 5)
    
    for face in faces:
        x, y ,w, h = face
        img  = cv2.rectangle(frame, (x, y), (x+w, y+h), (255,255,0),2)
    
    
    key_pressed = cv2.waitKey(1) & 0xFF
    
    if(key_pressed==ord('q')):
        break
        
    cv2.imshow("frame", frame)
    
cam.release()
cv2.destroyAllWindows()
```
